```python
import numpy as np
from sklearn.decomposition import FastICA
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
# generate data
data_y1 = np.linspace(-1,1,101)
data_x1 = np.linspace(0,0,101)
line1 = np.stack((data_x1, data_y1),1)
data_y2 = np.linspace(-1,1,101)
data_x2 = np.linspace(-1,1,101)
line2 = np.stack((data_x2, data_y2),1)

all_data = np.concat((line1,line2))
print(all_data.shape)
normalized = StandardScaler().fit_transform(all_data)
ica = FastICA(n_components=2, max_iter=1000).fit_transform(normalized)
"""plt.figure()
plt.scatter(all_data[:,0], all_data[:,1])
plt.title("original mixed data")
plt.show()

plt.figure()
plt.scatter(normalized[:,0], normalized[:,1])
plt.title("normalized mixed data")
plt.show()"""

print(ica.shape)
plt.figure()
plt.scatter(ica[:,0], ica[:,1])
plt.title("data after ica (randomly + or x)")
plt.show()
```
